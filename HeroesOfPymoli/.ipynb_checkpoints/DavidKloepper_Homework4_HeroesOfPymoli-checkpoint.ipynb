```python
#Import dependencies
import pandas as pd
import numpy as np

#Set file path of purchase data
filePath = "Resources/purchase_data.csv"

#Read in purchase data to a dataframe
purchaseData_df = pd.read_csv(filePath)
```

```python
purchaseData_df.head()
```

```python
#Count total players by unique screen names
totalPlayers = purchaseData_df["SN"].nunique()

#Put count of total players into a data frame
totalPlayers_df = pd.DataFrame({"Total Players":[totalPlayers]})

#Rename the index for cleaner printing
totalPlayers_df.index = ["Total Players"]

#Print total players dataframe
totalPlayers_df
```

```python
#Count unique items by item ID
uniqueItems = purchaseData_df["Item ID"].nunique()

#Calc average price of all items
avgPrice = purchaseData_df["Price"].mean()

#Count purchases by unique purchase IDs
totalPurchases = purchaseData_df["Purchase ID"].nunique()

#Calc total revenue
totalRevenue = purchaseData_df["Price"].sum()

#Put results into a new dataframe
summary_df = pd.DataFrame({"Number of Unique Items":[uniqueItems],"Average Price":[avgPrice],
                           "Number of Purchases": [totalPurchases],"Total Revenue": [totalRevenue]})

#Rename the index for cleaner printing
summary_df.index = ["Summary"]

#Format the currency columns
summary_df["Average Price"] = summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)

#Print the summary dataframe
summary_df
```

```python
#Set new dataframe, grouping by Gender field
genderStats_df = purchaseData_df.groupby("Gender")

#Count number of unique screen names by gender
genderPlayerCount = genderStats_df.nunique()["SN"]

#Calc percent of players by dividing player count for gender to total players calculated in earlier cell
genderPctPlayers = (genderPlayerCount / totalPlayers) * 100

#Create data frame with values for player count and percent. Gender is the index by default
genderDemo_df = pd.DataFrame({"Total Count": genderPlayerCount,"Percentage of Players": genderPctPlayers})

#Set the index name of the dataframe for a nice printout
genderDemo_df.index.name = "Gender"

#Format the percentage field to a percentage format with two decimal places
genderDemo_df["Percentage of Players"] = genderDemo_df["Percentage of Players"].map("{:,.2f}%".format)

#Print the dataframe
genderDemo_df
```

```python
#Using grouped dataframe from previous cell, count purchase IDs by gender 
genderPurchaseCount = genderStats_df["Purchase ID"].count()

#Calculate average purchase price by gender
genderAvgPrice = genderStats_df["Price"].mean()

#Calculate total purchase amounts by gender 
genderTotalPurchase = genderStats_df["Price"].sum()

# Average purchase total by gender divivded by purchase count by unique shoppers
genderAvgPurchase = genderTotalPurchase/genderPlayerCount

#Create data frame with values for purchase metrics by gender. Gender is the index by default
genderPurchases_df = pd.DataFrame({"Purchase Count": genderPurchaseCount, 
                                    "Average Purchase Price": genderAvgPrice,
                                    "Total Purchase Value":genderTotalPurchase,
                                    "Avg Total Purchase per Person": genderAvgPurchase})

#Set the index name of the dataframe for a nice printout
genderPurchases_df.index.name = "Gender"

#Format the fields with dollar amounts to a currency format with two decimals
genderPurchases_df["Average Purchase Price"] = genderPurchases_df["Average Purchase Price"].map("${:,.2f}".format)
genderPurchases_df["Total Purchase Value"] = genderPurchases_df["Total Purchase Value"].map("${:,.2f}".format)
genderPurchases_df["Avg Total Purchase per Person"] = genderPurchases_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

#Print the dataframe
genderPurchases_df
```

```python
#Setup 8 bins for age ranges
ageBins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 999]
binNames = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Segment ages into the bins above, using initial purchase data dataframe
purchaseData_df["Age Group"] = pd.cut(purchaseData_df["Age"],ageBins, labels=binNames)

#Create a new dataframe grouped by the newly binned age groups
ageGroups_df = purchaseData_df.groupby("Age Group")

#Count number of unique screen names by age group
ageTotalPlayers = ageGroups_df["SN"].nunique()

#Calc percent of players by dividing player count for gender to total players calculated in earlier cell
agePctPlayers = (ageTotalPlayers/totalPlayers) * 100

#Create data frame with values for player count and percent. Age Group is the index by default
ageDemos_df = pd.DataFrame({"Total Count": ageTotalPlayers, "Percentage of Players": agePctPlayers})

#Set the index name of the dataframe for a nice printout
ageDemos_df.index.name = "Age Ranges"

#Format the percentage field to a percentage format with two decimal places
ageDemos_df["Percentage of Players"] = ageDemos_df["Percentage of Players"].map("{:,.2f}%".format)

#Print the dataframe
ageDemos_df
```

```python
#Using grouped dataframe from previous cell, count purchase IDs by the binned age groups
agePurchaseCount = ageGroups_df["Purchase ID"].count()

#Calculate average purchase price by age group
ageAvgPrice = ageGroups_df["Price"].mean()

#Calculate total purchase amounts by age group
ageTotalPurchase = ageGroups_df["Price"].sum()

#Calculate average purchase total by age group divivded by purchase count by unique shoppers
ageAvgPurchase = ageTotalPurchase/ageTotalPlayers

#Create data frame with values for purchase metrics by gender. Age Range is the index by default
agePurchases_df = pd.DataFrame({"Purchase Count": agePurchaseCount,
                                 "Average Purchase Price": ageAvgPrice,
                                 "Total Purchase Value": ageTotalPurchase,
                                 "Avg Purchase Total per Person": ageAvgPurchase})

#Set the index name of the dataframe for a nice printout
agePurchases_df.index.name = "Age Ranges"

#Format the fields with dollar amounts to a currency format with two decimals
agePurchases_df["Average Purchase Price"] = agePurchases_df["Average Purchase Price"].map("${:,.2f}".format)
agePurchases_df["Total Purchase Value"] = agePurchases_df["Total Purchase Value"].map("${:,.2f}".format)
agePurchases_df["Avg Purchase Total per Person"] = agePurchases_df["Avg Purchase Total per Person"].map("${:,.2f}".format)

#Print the data frame
agePurchases_df
```

```python
#Group purchase data by screen names
spenderStats_df = purchaseData_df.groupby("SN")

#Calculate total purchases by screen name
spenderPurchaseCount = spenderStats_df["Purchase ID"].count()

#Calculate average price by screen name
spenderAvgPrice = spenderStats_df["Price"].mean()

#Calculate total purchase amounts by screen name
spenderTotalPurchase = spenderStats_df["Price"].sum()

#Create data frame with values for player count and percent. Screen name is the index by default
spenderPurchases_df = pd.DataFrame({"Purchase Count": spenderPurchaseCount,
                             "Average Purchase Price": spenderAvgPrice,
                             "Total Purchase Value":spenderTotalPurchase})

#Sort the dataframe descending by the total purchase value
spenderPurchases_df = spenderPurchases_df.sort_values(["Total Purchase Value"], ascending=False)

#Set the index name of the dataframe for a nice printout
spenderPurchases_df.index.name = "Top Purchasers (SN)"

#Format the fields with dollar amounts to a currency format with two decimals
spenderPurchases_df["Average Purchase Price"] = spenderPurchases_df["Average Purchase Price"].map("${:,.2f}".format)
spenderPurchases_df["Total Purchase Value"] = spenderPurchases_df["Total Purchase Value"].map("${:,.2f}".format)

spenderPurchases_df.head(5)
```

```python
#Create new dataframe with only the item-specific columns
items_df = purchaseData_df[["Item ID", "Item Name", "Price"]]

#Create new dataframe grouping the item data by item id and item name 
itemGroup_df = items_df.groupby(["Item ID","Item Name"])

#Count number of times price appears to get count of purchases
itemPurchaseCount = itemGroup_df["Price"].count()

#Calculate total dollars purchased by item
itemPurchaseTotal = (itemGroup_df["Price"].sum()) 

#Calculate actual item price dividing total amount purchased by number of items purchased
itemPrice = itemPurchaseTotal/itemPurchaseCount

#Create data frame with values for player count and percent. Item ID and Item Name are the indicies by default
popularItems_df = pd.DataFrame({"Purchase Count": itemPurchaseCount, 
                                   "Item Price": itemPrice,
                                   "Total Purchase Value":itemPurchaseTotal})

#Sort the dataframe descending by total number of purchases
popularItemsSort_df = popularItems_df.sort_values(["Purchase Count"], ascending=False)

#Format the fields with dollar amounts to a currency format with two decimals
popularItemsSort_df["Item Price"] = popularItems_df["Item Price"].map("${:,.2f}".format)
popularItemsSort_df["Total Purchase Value"] = popularItems_df["Total Purchase Value"].map("${:,.2f}".format)

#Put top five items into a new dataframe and print
topFiveItems = popularItemsSort_df.head(5)
topFiveItems
```

```python
# Take the popularItems_df items dataframe, sort by the total purchase price and put into a new dataframe
profitableItems_df = popularItems_df.sort_values(["Total Purchase Value"],
                                                   ascending=False)

#Format the fields with dollar amounts to a currency format with two decimals
profitableItems_df["Item Price"] = profitableItems_df["Item Price"].map("${:,.2f}".format)
profitableItems_df["Total Purchase Value"] = profitableItems_df["Total Purchase Value"].map("${:,.2f}".format)

#Put top five items into a new dataframe and print
mostProfitableItems = profitableItems_df.head(5)
mostProfitableItems
```
